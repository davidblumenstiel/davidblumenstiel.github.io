```python
#Imports

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator


import os
import shutil
#for dirname, _, filenames in os.walk('/kaggle/input'):
#    for filename in filenames:
#        print(os.path.join(dirname, filename))

```

```python
labels = pd.read_csv("../input/cassava-leaf-disease-classification/train.csv")
len(labels)
```

```python
#This makes new directories for training images, and copies to them.  This is needed for flow_from_directory

os.mkdir('/kaggle/images')
os.mkdir('/kaggle/images/train_images')
os.mkdir('/kaggle/images/train_images/class0')
os.mkdir('/kaggle/images/train_images/class1')
os.mkdir('/kaggle/images/train_images/class2')
os.mkdir('/kaggle/images/train_images/class3')
os.mkdir('/kaggle/images/train_images/class4')


for image_file in os.listdir('/kaggle/input/cassava-leaf-disease-classification/train_images'):
    
    source = '/kaggle/input/cassava-leaf-disease-classification/train_images/' + image_file #Get's the source directory for file
    
    label = int(labels[labels.image_id == image_file].label) #Returns the label for the file
    destination = '/kaggle/images/train_images/class' + str(label) + '/' + image_file
    
    shutil.copy(source, destination)
```

```python
image = cv2.imread("/kaggle/images/train_images/class0/1000015157.jpg",1)   #Will give you the array of pixel values
plt.imshow(image)
```

```python
train_datagen = ImageDataGenerator(validation_split=0.2,
                                rescale = 1/255,
                                rotation_range=20, 
                                zoom_range=0.15,
                                width_shift_range=0.2,
                                height_shift_range=0.2,
                                shear_range=0.15,
                                horizontal_flip=True,
                                fill_mode="nearest")

val_datagen = ImageDataGenerator(validation_split=0.2,
                                rescale = 1/255)


shape = (128,128,3)


train_generator = train_datagen.flow_from_directory('/kaggle/images/train_images', 
                                                    class_mode="categorical",
                                                    batch_size = 32, 
                                                    target_size = shape[0:2],
                                                    shuffle = True, 
                                                    subset="training")

validation_generator= val_datagen.flow_from_directory('/kaggle/images/train_images', 
                                                      class_mode="categorical",
                                                      batch_size = 32, 
                                                      target_size = shape[0:2],
                                                      shuffle = False, 
                                                      subset="validation")



files = os.listdir("/kaggle/images/train_images/")
len(files)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense

model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = (128,128,3)))
model.add(Conv2D(filters = 16, kernel_size = 3, activation = 'relu'))
model.add(MaxPooling2D((2, 2),padding='same'))

model.add(Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
model.add(Conv2D(filters = 16, kernel_size = 3, activation = 'relu'))
model.add(MaxPooling2D((2, 2),padding='same'))

model.add(Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
model.add(Conv2D(filters = 16, kernel_size = 3, activation = 'relu'))
model.add(MaxPooling2D((2, 2),padding='same'))

model.add(Flatten())
model.add(Dense(32, activation='linear'))
model.add(Dense(5, activation='softmax'))

model.summary()

model.compile(loss = 'categorical_crossentropy',
              optimizer = 'rmsprop',
              metrics = 'accuracy')
```

```python
history = model.fit_generator(train_generator,
        epochs=20,
        validation_data=validation_generator,
        use_multiprocessing=True,
        workers = 4)
```
